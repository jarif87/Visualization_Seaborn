```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("/content/best_cs_scientist_details.csv")
df.head()
```

# Lineplot

```python
plt.rcParams["figure.figsize"] = (12, 8)
sns.lineplot(x="National Rank", y="#DBLP",data=df)
plt.show()
```

#barplot

```python
plt.figure(figsize=(10,8))
country=df["Country"].value_counts().to_dict()
count_key=list(country.keys())
count_value=list(country.values())
sns.barplot(x=count_key,y=count_value)
plt.xticks(rotation=90)
plt.show()
```

# Histogram

```python
plt.figure(figsize=(10,7))
sns.histplot(data=df,x="#DBLP",kde=True)
plt.show()
```

# boxplot

```python
plt.figure(figsize=(10,7))
sns.boxplot(data=df[["#DBLP","H-Index"]])
plt.show()
```

# violin plot

```python
plt.figure(figsize=(10,7))
sns.violinplot(data=df[["H-Index","#DBLP"]],split=True, gap=.3, inner="quart")
plt.show()
```

```python

sns.relplot(data=df, x="#DBLP", y="National Rank",height=6)
plt.show()
```

# Bubble plot

```python
sns.relplot(data=df[: 100], x="Citations", y="#DBLP",height=6,size="World Rank",hue="Country")
plt.show()
```

# Subplots

```python
sns.relplot(data=df[: 100], x="Citations", y="#DBLP",height=20,size="World Rank",
            col='Country',hue="World Rank",col_wrap=2)
plt.show()
```

# Advance plots

# strip plot

```python
sns.catplot(data=df, x="#DBLP", y="Country",kind="strip",height=7,aspect=2)
```

```python
sns.catplot(data=df, x="#DBLP", y="Country",kind="box",height=7,aspect=2)
```

```python
sns.catplot(data=df, x="#DBLP", y="Country",kind="violin",height=7,aspect=2)
```

```python
sns.catplot(data=df, x="#DBLP", y="Country",kind="boxen",height=7,aspect=2)
```

# Density plot

```python
sns.kdeplot(data=df, x="#DBLP",fill=True)
plt.show()
```
