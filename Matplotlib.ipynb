```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("/content/best_cs_scientist_details.csv")
df.head()
```

```python
plt.rcParams["figure.figsize"]=(10,6)
plt.plot(df["World Rank"],df["#DBLP"])
plt.show()
```

```python
plt.rcParams["figure.figsize"]=(10,6)
plt.plot(df["World Rank"],df["#DBLP"])
plt.xlabel("world rank")
plt.ylabel("publications")
plt.show()
```

```python
plt.plot(df["#DBLP"])
plt.plot(df["H-Index"])
```

```python
plt.plot(df["#DBLP"][:20])
plt.plot(df["H-Index"][:20])
plt.legend(labels=["DBLP","H-Index"])
plt.xticks(ticks=df["World Rank"][:20],labels=df["Name"][: 20],rotation=90)
plt.show()
```

```python

n_rank=df["National Rank"][:30]
cita=df["Citations"][:30]
dblp=df["#DBLP"][:30]
world=df["World Rank"][:30]
name=df["Name"][:30]
fig,ax=plt.subplots(nrows=2,ncols=2)
ax[0,0].plot(n_rank,color="blue")
ax[0,1].plot(cita,color="green")
ax[1,0].plot(dblp,color="red")
ax[1,1].plot(world,color="darkorange")
ax[0,0].set_title("National rank")
ax[0,1].set_title("Citations")
ax[1,0].set_title("DBLP")
ax[1,1].set_title("Name")

for x in range(2):
  for y in range(2):
    ax[x,y].set_xticks(ticks=world)
    ax[x,y].set_xticklabels(labels=name,rotation=90)
plt.show()
```

```python
country_count=df["Country"].value_counts().to_dict()
```

```python
x=list(country_count.keys())
y=list(country_count.values())
plt.plot(y,color="crimson")
plt.xticks(ticks=range(len(x)),labels=x,rotation=90)
plt.show()
```

```python
x=list(country_count.keys())
y=list(country_count.values())
plt.bar(x,y,color="crimson")
plt.xticks(ticks=range(len(x)),labels=x,rotation=90)
plt.show()
```

```python
plt.hist(df["#DBLP"],bins=40,color="navy")
plt.show()
```

```python
outliers=dict(markerfacecolor="red",marker="P")
plt.boxplot(df["#DBLP"],flierprops=outliers)
plt.show()
```
