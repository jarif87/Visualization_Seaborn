```python
import pandas as pd
import plotly.express as px
# ! pip install -q plotly cufflinks
# from plotly.offline import init_notebook_mode
# init_notebook_mode(connected=True)
```

```python
df=pd.read_csv("/content/best_cs_scientist_details.csv")
df.head()
```

```python
 fig=px.line(df, x="World Rank", y="#DBLP", title='Data Visualize')
 fig.show()
```

```python
 fig=px.line(df, x="World Rank", y=["#DBLP","National Rank","H-Index"], title='Data Visualize')
 fig.show()
```

```python
fig = px.scatter(df, x="National Rank", y="Citations")
fig.show()
```

```python
c_count=df["Country"].value_counts().to_dict()
country=list(c_count.keys())
value=list(c_count.values())
fig = px.bar(x=country, y=value,labels={"x":"Country","y":"Count"})
fig.show()
```

```python
fig = px.histogram(df, x="#DBLP",color_discrete_sequence=['darkorange'])
fig.update_layout(title=dict(text="Histogram",x=0.5, y=0.95))
fig.show()
```

```python
fig = px.box(df, y="#DBLP",points="all")
fig.show()
```

```python
fig = px.box(df, x=["Citations","#DBLP"])
fig.show()
```

```python
fig=px.scatter(df, x="Citations", y="#DBLP",color="Country",symbol="Name",size="H-Index",hover_data=["World Rank","National Rank","Name","Affiliation"])
fig.show()
```

```python
fig = px.violin(df, y="#DBLP",box=True,points="all",hover_data=df.columns)
fig.show()
```
